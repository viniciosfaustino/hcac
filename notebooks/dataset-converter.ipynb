```python
import numpy as np
import pandas as pd
import csv
import pickle
import os
import re
from gensim.models import KeyedVectors
```

# Loading pretrained embeddings

```python
model = KeyedVectors.load_word2vec_format("../embeddings/skip_s50.txt")
```

# Function to save the pickle file

```python
def save_to_file(dataset, file_name):
    name = os.path.splitext(file_name)[0]
    with open(os.path.join("..",'datasets',str(name+'.data')), 'wb+') as file_handler:
        pickle.dump(dataset, file_handler)
        print("file saved:", name)
```

# Function that create and embedding vector for the sentence

```python
def create_embedding(dataset_name, _sep):
    file_path = os.path.join("..","datasets","text",dataset_name)
    df = pd.read_csv(file_path, sep=_sep, header=None, index_col=False)
    data = []
    for i, line in enumerate(df.values[:,0]):    
        sentence_vector = np.zeros(model.vector_size)        

        for word in line.split(' '):
            word = re.sub(r'[^\w\s]','',word)
            try:            
                sentence_vector += model.word_vec(word.lower())                
            except:
                pass    
        norm = np.linalg.norm(sentence_vector, axis=0)
        if norm > 0:
            sentence_vector = sentence_vector / norm    
        data.append(sentence_vector)
        
    labels = df.values[:,-1]
    labels = np.array(labels).reshape(len(data),1)
    dataset = np.append(data, labels, axis=1)    
#     dataset = np.array(dataset)
    
    return dataset
```

```python
name_sep = [("eleicao.csv", ';'), ("dilma.csv", '\t'), ("brasui.csv", '\t'), ("tweets.csv", '\t'), ("solo.csv", ',')]
for ns in name_sep:    
    dataset = create_embedding(ns[0], ns[1])
    save_to_file(dataset, ns[0])
```
